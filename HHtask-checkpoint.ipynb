```python
from neo4j import GraphDatabase
import pandas as pd
from transliterate import translit
```

Подключимся к локальной базе данных Neo4j Desktop

```python
uri="bolt://localhost:7687"
pwd="Crushing1nThirties"
driver=GraphDatabase.driver(uri, auth=("neo4j",pwd))
```

Загрузим датасет ( я заменил ";" на "," перед загрузкой, ибо pandas не мог различить колонки)

```python
df=pd.read_csv("data_test.csv")
df.head()
```

Есть подозрения, что некоторые имена в событиях повторяются, это будет означать, что существуют узлы с более чем одной связью

```python
f1=df["ФИО участника события 1"]
print(f"Кол-во уникальных значений в ФИО УЧАСТНИКА СОБЫТИЯ 1 {len(f1.unique())} и кол-во неуникальных значений {len(f1)}")
```

Изменим датасет так, чтобы было удобнее загружать его в базу данных, код неэффективный, поэтому занимает много времени, я запускал фоновым процессом, но в теории можно распаралелить

```python
f1u=f1.unique().tolist()
f2u=df["ФИО участника события 2"].unique().tolist()
fu=[]
fu.extend(f1u)
fu.extend(f2u)
fu=list(set(fu))

dct={}
for i in fu:
    dct[i]=[]
maxc=len(dct)*4930
c=0
for i in dct.keys():
    for index, row in df.iterrows():
        c+=1
        if i==row[1]:
            p=(row[0],row[2])
            dct[i].append(p)
        if i==row[2]:
            p=(row[0],row[1])
            dct[i].append(p)
    print(f"{100*c/maxc}% out of 100%")
print(dct)
```

```python
#query = "CREATE (:Movie {title: 'Batman Begins'})"
#query = """MATCH (n:Movie {title: 'Batman Begins'})
#DELETE (n)"""
#CREATE (a)-[r:RELTYPE {name: a.name + '<->' + b.name}]->(b)
#query="MATCH (n) DELETE(n)"

def fixname(tr_name):
    fixed_name=translit(tr_name, "ru", reversed=True)
    fixed_name=fixed_name.replace("'","_")
    fixed_name=fixed_name.replace(" ","_")
    return fixed_name

query="CREATE "
for i in dct.keys():
    tr_i=fixname(i)
    query+=f"({tr_i}:Person "+"{name: '"+i+"'}), \n"
for i in dct.keys():
    tr_i=fixname(i)
    for j in dct[i]:
        query+=f"({tr_i})-[:PARTICIPATED "+"{eventid:"+f"{j[0]}"+"}]->"+f"({fixname(j[1])}), \n"

query=query[0:len(query)-3]

print(query)
```

```python
with driver.session() as session:
    result = session.run(query)
    print(result.single())
```

Сделаем пробный запрос, на людей, которые учавствовали более чем в одном инциденте

```python
query="""MATCH ()-[PARTICIPATED]->(n)
WITH n, count(PARTICIPATED) as rel_cnt
WHERE rel_cnt > 1
RETURN n,rel_cnt"""
with driver.session() as session:
    result = session.run(query)
    pcnt=0
    for i in result:
        print(f"PERSON "+str(i[0]["name"])+" HAS PARTICIPATED IN "+str(i[1]) +" EVENTS")
        pcnt+=1
    print(f"AMOUNT OF PEOPLE WITH MORE THAN ONE CONNECTION IS {pcnt}")
```
